```
#ejercicio 3.-verificar si una persona es mayor e edad
"""
Analisis
Datos Entrada: edad - nombre
Proceso: mensaje("Es mayor de edad")
         mensaje("Es menor de edad")
"""
#codificacion
#Datos de entrada
nombre=input("ingrese su nombre:")
# Removed redundant initial input, now the loop handles all age inputs
while True:
  try:
    edad=int(input("ingrese su edad:"))
    if edad>=0 and edad<=100: # Corrected validation: age must be between 0 and 100 inclusive
      break
    else:
      print("Error: ingrese una edad valida de 0 a 100")
  except ValueError:
    print("Error: ingrese un numero valido")
#proceso
if edad>=18:
  print(f"la persona {nombre} es mayor de edad")
else:
  print(f"la persona {nombre} es menor de edad")
```

```
#ejercicio 4: Calcular el factorial de un número: Ejemplo el factorial del numero 5 es 120      ->5*4*3*2*1

# Pedir al usuario que ingrese un número
while True:
    try:
        num = int(input("Ingrese un número entero no negativo para calcular su factorial: "))
        if num < 0:
            print("Por favor, ingrese un número no negativo.")
        else:
            break
    except ValueError:
        print("Entrada inválida. Por favor, ingrese un número entero.")

# Calcular el factorial
factorial = 1
if num == 0:
    print(f"El factorial de 0 es 1")
elif num > 0:
    for i in range(1, num + 1):
        factorial = factorial * i
    print(f"El factorial de {num} es {factorial}")
```
